# Data Analytics with SAX

###### Below are several use cases of Java garbage collection (GC). Each case identifies a behavioral pattern over a period of time. The raw data in the use cases is simple time-series data with time on the x-axis and memory utilization on the y-axis. Each use case will be converted into its corresponding symbolic format (SAX) and used as a pattern for machine learning and use by a decision tree. The goal here being the ability to identify patterns over time and evaluate whether or not it is problematic. 

## Case 1 - Normal GC

###### This use case is a typical oscillating GC pattern where the memory climbs and is then periodically reclaimed. This behavior would be expected and should not result in error identification.

```python
import pandas as pd
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
from IPython.display import display

window_size = 10
alphabet = ['a','b','c','d','e','f','g','h']
alphabet_size = len(alphabet)

# SAX Boundaries

boundaries = norm.ppf(np.linspace(1./alphabet_size, 
                                  1-1./alphabet_size, 
                                  alphabet_size-1))

sax_dict = dict(zip(boundaries, alphabet))

case1 = pd.read_csv('PycharmProjects/saxanalytics/case1.csv', 
                    sep=',', 
                    parse_dates=[0], 
                    header=None, 
                    names=['Time','GC'])

# z-norm the data
case1 = ((case1 - case1.mean())/case1.std())

# discretize the data using piecewise aggregate approximation (PAA)
paa = np.array_split(case1['GC'], (len(case1['GC']) / window_size) + 1)
paa = np.asarray(map(lambda xs: xs.mean(), paa))
paa_size = len(paa) - 1

ax = case1.plot()
ax.set_ylabel('Memory (MB)')

x1 = 0
x2 =  window_size

for i in xrange(0, paa_size):
    ax.plot([x1, x2], [paa[i], paa[i]], '-k')
    ax.plot([x2, x2], [paa[i], paa[i+1]], '-k')
    x1 = x1 + window_size
    x2 = x2 + window_size
    
for line in boundaries:
    ax.plot([0,len(case1)],[line,line])

sax_data = list()

# Convert to SAX representation
for value in paa:
    for key in sort(sax_dict.keys()):
        if value < key:
            sax_data.append(sax_dict[key])
    sax_data.append(alphabet[len(alphabet) - 1])
     
display(''.join(sax_data))
```

## Case 4 - Memory Leak (Increasing Memory Utilization)

###### In this use case we see an increasing memory utilization where the lower boundary is climbing. This is indiciative of a memory leak (or some data otherwise not being reclaimed). Prior to crashing the application will experience performance impact while the virtual machine attempts to reclaim memory with increasing frequency.

## Case 5 - Irregular GC (Not Error Condition)
